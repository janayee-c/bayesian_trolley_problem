## The Trolley Problem

The ‘Trolley Problem’ is a series of thought experiments that have long intrigued ethicists, psychologists, and AI researchers alike. Originally presented in 1967 by philosopher Philippa Foot and later popularised by fellow philosopher Judith Jarvis Thomson in 1976, the Trolley Problem poses a moral dilemma in the context of a runaway trolley. The scenario forces us to consider the monumental decision of whether to sacrifice one person to save many others.

<br>

At the heart of the Trolley Problem is the “switch” scenario. In this setup, a trolley is hurtling down a track towards five people who are unable to move. A bystander, standing by a lever, has the power to alter the trolley’s course. By pulling the lever, the trolley would be diverted onto a sidetrack, sparing the five people but resulting in the death of one person who is on the alternate track. The moral quandary posed is stark: should the bystander do nothing and allow the trolley to kill the five people, or should they intervene, sacrificing one to save many?

<br>

This ethical puzzle is not just theoretical but taps into deeper questions about human morality and decision-making under pressure. It challenges us to articulate and justify our moral intuitions and to examine how contextual details might influence our judgments. The Trolley Problem continues to be a vital tool in Cognitive Science, providing insights into the principles that govern our ethical reasoning, moral psychology, and the values we hold most dear.


#Article - Moral Learning: Conceptual foundations and normative relevance

Moral assessments are increasingly recognized as exhibiting variability dependent on the context. An expanding body of research has pinpointed various contextual elements (such as emotions and intentions) that can systematically sway moral judgments. However, synthesizing these diverse influences into a cohesive framework for understanding moral judgments poses a challenge. Railton (2017) recently endeavored to tackle this challenge by proposing a causal-evaluative modeling approach to moral judgment. Supporting this model, Railton offers evidence from new variations of classic moral dilemmas akin to the trolley problem.

<br>

In this essay, we will be exploring how Railton's paper explores foundational concepts of moral learning through these variations of the trolley problem; the author reasons that moral judgement is largely determined by both model-based and model-free learning processes that can accurately predict responses to ethical dilemmas and circumstantial moral perception.

<br>

*Model-based learning* occurs when humans construct and update cognitive models that structure causal and evaluative information based on feedback loops from the environment. These models evolve with new experiences and allow individuals to extend their moral reasoning beyond emotion or intuition --largely through 'imaginative projection' that predicts hypothetical scenarios and simulative empathy based on social interaction.

<br>

On the other hand, *model-free learning* relies on implicit cues that are taken from experiences that have formed pre-existing cognitive models which subsequently guide automatic responses without active deliberation. Model-free processes depend on stored 'caches' of expectation values or individual actions that create immediate, emotion-driven moral responses. Theoretically, this allows humans to make decisions intuitively without reasoning through the specifics of every circumstance, resulting in a moral judgement that is often difficult to explain from a 'logical' perspective.

<br>

Railton illustrates how this interplay between the two learning types can result in asymmetric responses through the classic Switch scenario -- a variation of the sacrifical trolley dillemma questioning whether a participant would pull a lever to divert an oncoming trolley to collide into a single person, resulting in their death, in lieu of allowing the trolley to kill 5 other people. Railton derives two subvariants of the sacrificial Switch scenario, each featuring a specific lethal action that results in the death of one person to'save' five other people in the Switch and Beckon contexts.

<br>

Despite the similarities between each scenario, Railton encountered asymmetrical results in the distribution of students who answered 'yes' or 'no' to performing the lethal action depending on whether the the act was committed through pulling a lever, waving, or making a beckoning motion.

## Railton's Theory

Railton theorizes that the asymmetric results of Switch, Beckon and Wave trolley cases are indicative of how each scenario impacts the process of model-free learning that ultimately impact the probability of committing lethal act. In terms of outcome, each of the three circumstances features the same consequence of 1 death for 5 lives; however, the specific hand action used within each scenario invokes different 'cached' expectation values that trigger distinct and intuitive moral judgements.  

<br>

In Switch, where pulling a lever leads to diverting the trolley, the absence of direct or indirect negative experiences with a physical lever leads to exclusion of a model-free intuitive evaluation of the situation; instead of relying on 'cached' expecttion values, the individual resonably traces the consequences and conciously evaluates models such as Utilitarianism. This results in a majority approval of the action in Switch.

<br>

Conversely, in Beckon, where beckoning a person leads to their death, the model-based learning process evaluates the act with a negative 'cached' expectation value. The physical act of gesturing to beckon implies that the person possesses higher intentionality and purpose in luring them to their death or perceiving the victim as a means to an end, leading to disapproval from the majority.

<br>

In Wave, where waving signals safety to others, the model-free learning process, qualifies as a cached positive valuatio and  influences the majority to approve of the action. These results demonstrate how the degree of model-free learning processes influences moral judgments in different scenarios.

## Simulating - Railton Data


```python
import numpy as np
import pandas as pd

def simulate_rail(n, a_b, a_s, a_w):
    beckon = []
    wave = []
    switch = []

    for i in range(n):
        outcome_b = np.random.choice([1,0], p=[a_b, 1-a_b])
        beckon.append(outcome_b)

        outcome_s = np.random.choice([1,0], p=[a_s, 1-a_s])
        switch.append(outcome_s)

        outcome_w = np.random.choice([1,0], p=[a_w, 1-a_w])
        wave.append(outcome_w)

    return beckon, switch, wave

# Probabilities for each scenario: 42% yes for Beckon, 86% yes for Switch, 87% yes for Wave
beckon_responses, switch_responses, wave_responses = simulate_rail(100, 0.42, 0.86, 0.87)


beckon_col = [-1] * 100  # Identifies the Beckon scenario (negative model-free)
switch_col = [0] * 100   # Identifies the Switch scenario (neutral)
wave_col = [1] * 100     # Identifies the Wave scenario (positive model-free)

# Combine data
response = beckon_responses + switch_responses + wave_responses
scenario = beckon_col + switch_col + wave_col
data = list(zip(response, scenario))

# Create DataFrame
df = pd.DataFrame(data, columns=['response', 'scenario'])

print(df)
```

## Mediator - Logistic Regression Model

In this study, we employ a logistic regression model to explore Railton's scenario, focusing specifically on how model-free processes in each specific scenario influence expectations of negative and positive outcomes --overall impacting the likelihood of committing the lethal act per person. We use the logistic function to convert the linear combinations of our predictors into log-odds, which provide a nuanced view of the probability shifts. This approach allows us to examine the direction and magnitude of changes in the probability of committing a lethal act in the switch, beckon and wave scnearios based on the presence and vector of model-free processes. Our analysis includes generating trace graphs and calculating posterior means for each parameter to quantitatively assess these effects.


###Prior Predictive analysis

In this analysis, the predictive graph demonstrates that the priors chosen for our model encapsulate significant uncertainty across each scenario. This feature indicates a well-mixed model that aptly reflects the inherent complexities of moral judgment situations. Specifically, in the beckon scenario, our prior simulation reveals a notable density of values at probabilities near 0 and 1. This density suggests a high degree of variance and range in our prior distribution, resulting in clustering at the extremes; simultaneously, the substantial variability and shifting slopes between these points highlight our limited preliminary understanding of the impact of model-free learning. High degree of variance is also observable in both the wave and switch scenarios, where the dispersion of curves distinguishes uncertainty. Overall, significant differences between the three scenarios cannot be inducted from the below graphs.

```python
def invlogit(linear_func):
    return np.exp(linear_func) / (1 + np.exp(linear_func))

with pm.Model() as logistic_regression:
    # Priors
    a1 = pm.Normal('beckon',0,3)
    a2 = pm.Normal('switch',0,3)
    a3 = pm.Normal('wave',0,3)
    b = pm.Normal('model_free_effect', 0, 1.5)

    p_b = pm.math.invlogit(a1 + (b*-1))
    p_s = pm.math.invlogit(a2 + (b*0))
    p_w = pm.math.invlogit(a3 + (b*1))

    # Likelihoods
    y1 = pm.Binomial('y1', n=1, p=p_b, observed=beckon_responses)
    y2 = pm.Binomial('y2', n=1, p=p_s, observed=switch_responses)
    y3 = pm.Binomial('y3', n=1, p=p_w, observed=wave_responses)

    # Sampling

    trace1 = pm.sample()
```

```python
# prior predictive simulation for logistic regression for below model

# Sample from the priors
a_1samples, a_2samples, a_3samples, b_samples = pm.draw([a1, a2, a3, b], draws=50)

# Define the inverse logit function
def invlogit(linear_func):
    return np.exp(linear_func) / (1 + np.exp(linear_func))

# Define x values
x_grid = np.linspace(-1, 1, 50)

# Define function for prior simulation
def prior_sim(a):
    for i in range(50):
        y_predict = invlogit(a[i] + b_samples[i] * x_grid)
        plt.plot(x_grid, y_predict)

# Create grid of plots
plt.figure(figsize=(15, 5))

# Plot prior simulations for each set of samples
plt.subplot(1, 3, 1)
prior_sim(a_1samples)
plt.title('Prior Simulation for Beckon (a1)')

plt.subplot(1, 3, 2)
prior_sim(a_2samples)
plt.title('Prior Simulation for Switch (a2)')

plt.subplot(1, 3, 3)
prior_sim(a_3samples)
plt.title('Prior Simulation for Wave (a3)')

plt.tight_layout()
plt.show()
```

##Case Two (Bad) - Accounting for Noise and Inconsequential 'Model-Free (+/-)' variable

For our bad case, we are suggesting that Railtons approach doesn't account for clear socio-cultural sources of noise. Further, we also claim that the causal pathway from Scenario and Model-Free is inconsequential.

### Systematic Oversights in Railton's Experiment:
The need to account for noise here may not seem explicit at first, however, when examining Railton's experiment we can see a few systematic oversights.

<br>

  - Firstly, Railton's sample population is exclusively students in his ethics class. This means that there is likely some form of social/political homogeneity which increases the degree of social conformity. Thus, a source of noise when generalizing can be 'Sampling Variability'. This refers to the changes in data that might occur simply because of the particular sample of individuals chosen. If the sample doesn't represent the broader population (as noted with the homogeneity issue), the findings might not generalize well, introducing variability when trying to apply conclusions more broadly.

  <br>

  - Secondly, the impact of personal and social dynamics, such as a participant’s desire to conform to perceived social norms or their psychological state at the time of the experiment, can also introduce noise. These factors might cause them to respond in ways that do not accurately reflect their true moral judgments.

    - This can include factors such as mood, irrability, ongoing events, etc.

<br>

In essence, we are trying to show that there are complex social factors adding noise to Railton's simple causal model. This makes intuitive sense as we would expect there to be discrepancies in how participants 'truly' percieve model-free (negative/positive). Another naive example could be that people from different cultures percieve the communicative  acts of beckoning and waving slightly different from others.

<br>

The reason for not making an explicit 'social conformity' variable is because of the anonymity of the Railton experiment. Since students are blocked from observing each others responses we do expect the issues we mentioned above to be substantial factors, however, not accounting for them at all is something we contend with. Especially in the case of these 'Trolley Problem' variations, one would at least expect innate differences in culture to affect. These effects have been observed in other research pertaining to the Trolley Problem.

<br>

A recent study suggested that moral responses to the trolley problems can vary significantly according to cultural backgrounds (Gold et al., 2023). These cultural differences were more pronounced when the consequences were less severe than death. Specifically, fewer British participants were hesitant to act in the standard Switch scenario, reflecting a possibly stronger inclination towards utilitarian principles that emphasize the greatest good for the greatest number. Conversely, Chinese participants demonstrated greater reluctance to sacrifice one person for the benefit of many, likely influenced by Confucian values that stress the sanctity of each individual life and the moral implications of directly causing harm. This divergence underscores the need to consider cultural backgrounds in moral psychology, as these foundational beliefs significantly shape moral reasoning and ethical behavior.

<br>

While the role of culture makes a good causal case, for this exploration, we also omit explicit "cultural" variables. We can justify this for several reasons:

<br>

- Firstly, culture is inherently complex and dynamic, encompassing diverse values, beliefs, and practices that are challenging to encapsulate within a single quantifiable variable without oversimplifying these elements. Accurately measuring such a multifaceted construct poses significant difficulties, as there is no standardized metric for assessing cultural variables across different populations. Including culture could also risk reinforcing stereotypes and promoting simplistic cultural narratives if not handled with nuance and sensitivity.

<br>

- Moreover, the causal relationships between culture and other study variables are often unclear, complicating the interpretation of data and potentially leading to erroneous conclusions about cultural influences on behavior. Researchers instead often choose to focus on more specific, directly observable variables that can offer more granular insights into cultural impacts, such as specific social norms or community values. This approach not only enhances the control and validity of experimental designs but also adheres to ethical standards by protecting participant anonymity and avoiding unnecessary categorization that could lead to bias or discrimination.

<br>

Hence, while the study of cultural effects is undeniably valuable, the decision to exclude culture as an explicit variable is frequently driven by a pursuit of methodological rigor and ethical research practice. # Simulating Data with Noise

```python
def simulate_rail_with_noise(n, a_b, a_s, a_w, noise_level):
    beckon = []
    wave = []
    switch = []

    for i in range(n):
        noisy_b = max(0, min(1, a_b + np.random.normal(0, noise_level)))
        outcome_b = np.random.choice([1,0], p=[noisy_b, 1-noisy_b])
        beckon.append(outcome_b)

        noisy_s = max(0, min(1, a_s + np.random.normal(0, noise_level)))
        outcome_s = np.random.choice([1,0], p=[noisy_s, 1-noisy_s])
        switch.append(outcome_s)

        noisy_w = max(0, min(1, a_w + np.random.normal(0, noise_level)))
        outcome_w = np.random.choice([1,0], p=[noisy_w, 1-noisy_w])
        wave.append(outcome_w)

    return beckon, switch, wave

# Simulate with added noise level of 0.1
n_beckon_responses, n_switch_responses, n_wave_responses = simulate_rail_with_noise(100, 0.42, 0.86, 0.87, 0.1)
```
